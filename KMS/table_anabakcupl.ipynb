```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import Dataset, DataLoader, random_split
```

```python
totaldf.shape
```

```python
DEVICE = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
```

```python

class MyDataset(Dataset):
    def __init__(self, df, target = 'tier_info.tier'):
        super(MyDataset, self).__init__()
        self.df = df
        self.yt = pd.get_dummies(self.df[target])
        # self.df['tier_info.tier'] = self.df['tier_info.tier'].astype('category').cat.codes
        # y = self.df[target]
        self.x = torch.tensor(self.df.drop([target], axis=1).to_numpy(dtype = np.float64), dtype=torch.float32).to(DEVICE)
        self.y = torch.tensor(self.yt.to_numpy(dtype=np.float64), dtype=torch.float32).to(DEVICE)
    
    def __len__(self):
        return self.df.shape[0]
    
    def __getitem__(self, idx):
        return self.x[idx], self.y[idx]

    def get_cols(self):
        return self.yt.columns.tolist()
```

```python
df = totaldf.sample(frac=0.1)
dataset = MyDataset(df)
df['tier_info.tier'].nunique()
# df = df[['rank',
#  'wards_killed_0',
#  'summoner.most_champions.win',
#  'wards_killed_1',
#  'summoner.level',
#  'wards_killed_2',
#  'vision_wards_bought_in_game_1',
#  'vision_wards_bought_in_game_0',
#  'vision_wards_bought_in_game_2',
#  'win']]
# dataset = MyDataset(df, 'rank')
```

```python
train_size = 0.7
val_size = 0.2
test_size = 0.1

seed = torch.manual_seed(42)
train_dataset, val_dataset, test_dataset = random_split(dataset, [train_size, val_size, test_size], seed)

trainDL = DataLoader(train_dataset, batch_size=64, shuffle=True)
valDL = DataLoader(val_dataset, batch_size=64, shuffle=True)
testDL = DataLoader(test_dataset, batch_size=64, shuffle=True)
```

```python
class MyModel(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(MyModel, self).__init__()
        self.seq = nn.Sequential(
            nn.Linear(input_dim, input_dim * 4),
            nn.Tanh(),
            nn.Linear(input_dim * 4, input_dim * 2),
            nn.Tanh(),
            nn.Linear(input_dim * 2, output_dim),
        )
        for model in self.seq:
            if isinstance(model, nn.Linear):
                nn.init.xavier_uniform_(model.weight)
    
    def forward(self, x):
        return self.seq(x)

class MyBatchModel(nn.Module):
    def __init__(self, input_dim, output_dim, active = nn.ReLU):
        super(MyBatchModel, self).__init__()
        self.seq = nn.Sequential(
            nn.Linear(input_dim, input_dim * 4),
            active(),
            nn.Tanh(),
            nn.Linear(input_dim * 4, input_dim * 2),
            active(),
            nn.Tanh(),
            nn.Linear(input_dim * 2, output_dim),
        )
        for model in self.seq:
            if isinstance(model, nn.Linear):
                nn.init.xavier_uniform_(model.weight)
    
    def forward(self, x):
        return self.seq(x)
```

```python
from torchmetrics.classification import MultilabelAccuracy, MultilabelF1Score
from torchmetrics.regression import MeanSquaredError, R2Score

model = MyModel(df.shape[1] - 1, df['tier_info.tier'].nunique()).to(DEVICE)
modelBatch = MyBatchModel(df.shape[1] - 1, df['tier_info.tier'].nunique(), nn.BatchNorm1d).to(DEVICE)

criterion = nn.CrossEntropyLoss()
evaluator = MultilabelF1Score(num_labels=df['tier_info.tier'].nunique())
# criterion = nn.HuberLoss()
# evaluator = MeanSquaredError()
optimizer = optim.Adam(model.parameters(), lr=0.01)
sched = torch.optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.1)


model
```

```python


torch.autograd.set_detect_anomaly(True)

def train(model, loader, criterion, optimizer):
    model.train()     
    for x, y in loader:
        
        ypred = model(x)
        loss = criterion(ypred, y)       
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

def train_with_l2(model, loader, criterion, optimizer, l2_lambda=0.001):
    model.train()
    for x, y in loader:
        
        ypred = model(x)
        loss = criterion(ypred, y)       
        
        l2_norm = sum(p.pow(2.0).sum() for p in model.parameters())
        loss = loss + l2_lambda * l2_norm
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

def test(model, trainloader, loader, loadername, criteria, evaluator=evaluator):
    acc = evaluator.to(DEVICE)
    model.eval()
    
    def calc_acc_loss(model, loader, criteria):
        accuracyList = []
        lossList = []
        for x, y in loader:
            ypred = model(x)
            loss = criteria(ypred, y)         
            
            accuracyList.append(acc(ypred, y).item())
            lossList.append(loss.item())
        return accuracyList, lossList
    
    
    with torch.no_grad():
        accuracyList = []
        accuracyList, lossList = calc_acc_loss(model, loader, criteria)
        accuracytrList, losstrList = calc_acc_loss(model, trainloader, criteria)

        final_traccu, fianl_trloss, final_accu, final_loss = np.mean(accuracytrList), np.mean(losstrList), np.mean(accuracyList), np.mean(lossList)

        print(f'Train f1 = {final_traccu} \nTrain loss = {fianl_trloss}')
        print(f'{loadername} f1 = {final_accu} \n{loadername} loss = {final_loss}')    

        return final_traccu, fianl_trloss, final_accu, final_loss   

# def predict(model, x, labels = dataset.get_cols()):
#     model.eval()
#     predictions = []
#     with torch.no_grad():
#         ypred = model(x)
#         ypred_loc = ypred.argmax(dim=1)
    
#     for x  in ypred_loc:
#         print(x)


```

```python
def run(EPOCONS = 20 , model = model):
    total_result = [[],[],[],[]]
    for epoch in range(EPOCONS):
        train(model, trainDL, criterion, optimizer)
        print(epoch, 'passed')
        result = test(model, trainDL, testDL, 'Test' , criterion, evaluator)
    
        for idx, tmpresult in enumerate(total_result):
            tmpresult.append(result[idx])
        if np.array(total_result[-2]).argmax() == epoch:
            torch.save(model, f'model_best_{model._get_name()}_epo_{epoch}.pth')

    return total_result

total_result = run()
total_batchResult = run(20, modelBatch)
```

```python
# final_traccu, fianl_trloss, final_accu, final_loss
import matplotlib as mpl
# 맑은 고딕
mpl.rc('font', family='Malgun Gothic')

colname = ['Train f1', 'Train loss', 'Test f1', 'Test loss']
def plot_result(total_result, title = 'result'):
    
    df_result = pd.DataFrame(dict(zip(colname, total_result)))
    ax = plt.subplot(111)
    ax.plot(df_result[colname[0]], label = colname[0], alpha= 0.3)
    ax.plot(df_result[colname[2]], label = colname[2], color='red')
    axtwin = ax.twinx()
    axtwin.plot(df_result[colname[1]], label = colname[1], alpha= 0.3)
    axtwin.plot(df_result[colname[3]], label = colname[3], color='pink')
    lines, label1 = ax.get_legend_handles_labels()
    lines2, label2 = axtwin.get_legend_handles_labels()
    axtwin.legend(lines + lines2, label1 + label2, loc=0)
    plt.title(title)
    plt.show()

plot_result(total_result, '일반 결과')
plot_result(total_batchResult, 'Batch 결과')
```
