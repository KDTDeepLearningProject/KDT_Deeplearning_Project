```python
import pandas as pd
import orjson 
import dask.dataframe as dd
import os
```

```python
SKIP = 3
total = pd.DataFrame()
# total_dask = dd.Dataframe()
result = [x for x in os.listdir("./json/") if x.endswith(".json")]
result.sort(key = lambda x: int(x.split(".")[0].split("_")[1]))
result = [x for x in result if (int(x.split(".")[0].split("_")[1]) % 3) == 2]

for idx, x in enumerate(result):
    x = './json/' + x
    if idx % SKIP == 0:
        with open(x, 'rb') as f:
            data = orjson.loads(f.read())
            tmp = pd.json_normalize(data['pageProps']['data'])
            if total.empty:
                total = tmp
            total = pd.concat([total, tmp], ignore_index=True, axis=0)
            # tables.append(data['pageProps']['data'])
            print(x)

        # break
        
```
